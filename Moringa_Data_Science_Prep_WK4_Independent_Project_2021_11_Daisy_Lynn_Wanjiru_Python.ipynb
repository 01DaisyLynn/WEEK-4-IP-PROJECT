**DATA CLEANING**

*This done by following the data integrity rules i.e Validity, Accuracy, Completeness, Consistency, Uniformity*




Validity

```python
#dropping/deleting
#we will drop the cars column since it contains data similar to Bluecar counter
dataset_df.drop(["Cars"], axis = 1, inplace = True) 
dataset_df.head() 
```

```python
#combining year	month	day	to form one column
dataset_df['Date'] = dataset_df[['year', 'month', 'day']].astype(str).agg('-'.join, axis=1)
dataset_df.head(2)
```

```python
#combining hour	minute to form one column
dataset_df['Time'] = dataset_df[['hour', 'minute']].astype(str).agg(':'.join, axis=1)
dataset_df.head(2)
```

Completeness

```python
#finding the sum of missing values in each column
dataset_df.isnull().sum()
```

```python
#using 0 to replace the missing values
#this particular method was used so that it won't intefere with the data/analysis
dataset_df['Scheduled at'].fillna(0, inplace=True)
dataset_df['Displayed comment'].fillna(0, inplace=True)

dataset_df.head(3)
```

Consistency

```python
#dropping duplicated data incase of any
dataset_df.drop_duplicates().tail()
```

Uniformity

```python
df=dataset_df.rename(columns={'Bluecar counter':'Bluecar'})
df.head(2)
```
